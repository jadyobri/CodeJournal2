**Uniform Histogram**

```python
#imports for numpy and matplotlib usage
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#set n to 1000 for 1000 random numbers between 0 and 1
n = 1000

#Creates uniform random between 0 and 1
x = np.random.uniform(low=0.0, high=1.0, size=n)

#plotting for histogram, idea from professor
#bins set to 100 for the 100 bins, inspired from:https://stackoverflow.com/questions/33458566/how-to-choose-bins-in-matplotlib-histogram
plt.hist(x, bins = 100,alpha = 0.5, edgecolor = 'black')

# labeling x axis as x
plt.xlabel('x')

#labeling y axis as y
plt.ylabel('y')

#titling the histogram as Uniform Histogram
plt.title('Uniform Histogram')
plt.savefig("prompt9.pdf", dpi = 300, bbox_inches="tight")
```
