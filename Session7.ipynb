**Range of 100 value from 0 to 1**

```python
#import mathplotlib.pyplot as plt for ploting later
import matplotlib.pyplot as plt

#numpy for linspace
import numpy as np

#ranger function takes linspace from 0.0 to 1.0 with 100 values and returns it
def ranger():
    x = np.linspace(0.0,1.0,num = 100)
    return x
```

**Exp(x)**

```python
#we take e from math
from math import e

#exp(x) is e^x
#so we take in x and find the e**x and return that as a float
def exp(x):
    y = float(e**x)
    return y
```

**Ploting the graph of x vs exp(x)**

```python
#we create x as ranger function and y as an empty list
#we then do for loop in all of the members of x and find the exp of them
#then we plot x vs exp(x) with plt.plot
#we also add x and y labels with plt.xlabel() and plt.ylabel() Functions to add "Time [milliseconds]" as the x label and "Awesomeness" as the y label
def main(): 
    x = ranger()
    y = []
    for i in x:
        y.append(exp(i))
    plt.plot(x,y)
    plt.xlabel("Time [milliseconds]")
    plt.ylabel("Awesomness")
    plt.savefig("prompt7.pdf", dpi = 300, bbox_inches="tight")
if __name__ == "__main__":
    main()
```
